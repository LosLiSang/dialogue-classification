```python
def read_datas(filename: str) -> list: 
    with open(filename, mode="r") as f:
        lines = f.readlines()
    return lines
        

def spilit_datas(lines: list) -> tuple:
    data_len = len(lines)
    dev_list, train_list, test_list = lines[:data_len//9 * 2], lines[data_len//9 * 2 + 1:data_len//9 * 5], lines[data_len//9 * 5 + 1:]
    return dev_list, train_list, test_list


def save_datas(data: list, filename: str):
    with open(filename, mode='w') as f:
        f.writelines(data)
```

```python
lines = read_datas('data_code.txt')
lines
```

```python
for data, filename in zip(spilit_datas(lines), ["dev.txt", "test.txt", "train.txt"]):
    save_datas(data, filename)
```
